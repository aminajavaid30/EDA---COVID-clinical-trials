### 5. Make 5 questions and solve them by using group by

#### Q1: What is the total number of enrollments for each status type?   

```python
# Enrollments for each individual status
df.groupby('Status')[['Enrollment']].sum()
```

#### Q2. What are the average enrollments by gender?

```python
# Average enrollments by gender
df.groupby('Gender')[['Enrollment']].mean()
```

#### Q3. What are the total number of enrollments by phases?

```python
# Enrollments by phases
df.groupby('Phases')[['Enrollment']].sum()
```

#### Q4. What are the total number of enrollments by study results?

```python
# Total enrollments by study results
df.groupby('Study Results')[['Enrollment']].sum()
```

#### Q5. What are the maximum enrollments by study type?

```python
# Maximum enrollments by study type
df.groupby('Study Type')[['Enrollment']].max()
```
